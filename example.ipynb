# Reading MILES SSP models

**Note**: for import reasons, this notebook can't be run whilst it is in the same directory as read_miles.py. Move it elsewhere to run it.

The [MILES simple stellar population models (SSPs)](http://miles.iac.es/pages/ssp-models.php) are a tabulated set of SSP models. This is a small package to read them.

```python
import numpy as np
import matplotlib.pyplot as plt
from read_miles import read_miles
```

```python
ssps = read_miles.MilesSSPs()
print('number of ages = ', ssps.nt)
print('number of metallicities = ', ssps.nz)
print('total number of templates = ', ssps.nt*ssps.nz)
print('size of wavelength array = ', ssps.lmd.shape)
```

The templates are stored in an array $X$

```python
print('tempalate array X has dimension = ', ssps.X.shape)
```

A specific template can be accessed as follows

```python
idx = 16
print(f'Properties of template {idx}:')
print('   age = ', ssps.t[idx])
print('   metallicity = ', ssps.z[idx])
plt.plot(ssps.lmd, ssps.X[:, idx])
```
